```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Reading a CSV file
data = pd.read_csv("/content/amazon.csv")
data.head()
```

```python
# Handling any missing values from dataset
data.isnull().sum()
data.dropna(inplace=True)
```

```python
vector = CountVectorizer()
features = vector.fit_transform(data["review_content"])
```

```python
similarity_mat = cosine_similarity(features)
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
data["product_id"] = encoder.fit_transform(data["review_id"])
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
data["discounted_price"] = StandardScaler().fit_transform(data["discounted_price"])
```

```python
def recommend_products(product_id, top_n=5):
    similar_products = np.argsort(similarity_matrix[product_id])[1:top_n+1]
    recommended_products = data.iloc[similar_products][["product_name", "actual_price","rating"]].values.tolist()
    return recommended_products
```

```python
target_product = data[data["product_name"] == "Flix Micro Usb Cable For Smartphone (Black)"].index[0]
```

```python
recommend = recommend_products(target_product)
print(f"Top 10 recommended products for The Customers:\n")
for i, products in enumerate(recommend, start=1):
    product_name, actual_price,rating = products
    print(f"{i}. {product_name} :  {actual_price}  - {rating}")
```
